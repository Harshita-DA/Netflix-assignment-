# Released Movie Over the years?

```python
df.head()
```

```python
movie = df[df.type == 'Movie']
movie_year = movie.groupby('release_year').size().reset_index(name = 'count')
print(movie_year)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(movie_year['release_year'], movie_year['count'], marker='o', linestyle='-')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

# Movies made on year basis?

```python
movie = df[df.type == 'Movie']
movie_year = movie.groupby('release_year').size().reset_index(name = 'count')
print(movie_year)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(movie_year['release_year'], movie_year['count'], marker='o', color='r')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='release_year', y='count', data=movie_year, palette='viridis')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)  
plt.show()
```

# show Top 10 Directors who gave the highest number of tv shows and movies to netflix?

```python
content = df['director'].value_counts().head(10)

sns.barplot(y = content.index, x = content.values)
plt.title("Top 10 Directors Producing Most Netflix Content")
plt.show()
```

# in how many movies / tv shows , tom cruise was cast?

```python
tom_cruise_movies_shows = df[df['cast'].str.contains('Tom Cruise', case=False, na=False)]

count = tom_cruise_movies_shows.shape[0]

print(f"Tom Cruise was cast in {count} movies/TV shows.")
```

# how many movies got the Tv-14 rating in the canada ?

```python
tv14_canada_movies = df[(df['type'] == 'Movie') & 
                             (df['rating'] == 'TV-14') & 
                             (df['country'] == 'Canada')]

count = tv14_canada_movies.shape[0]

print(f"Number of movies with TV-14 rating released in Canada: {count}")
```

```python
rating_counts = df[(df['type'] == 'Movie') & (df['country'] == 'Canada')]['rating'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index,rotatelabels = True,  autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Distribution of Movie Ratings in Canada')
plt.show()
```
